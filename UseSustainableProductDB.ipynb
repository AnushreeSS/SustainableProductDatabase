# Using Sustainable Product Data

```python
import sqlite3
import pandas as pd
import gzip
```

```python
DB_FILE_PATH = "sustainable_product_db_final.db"
TABLE_NAME = "V_SUSTAINABILITY_PRODUCT_INFO"
# Unzip the compressed file
with gzip.open("sustainable_product_db_final.db.gz", "rb") as f1:
    with open(DB_FILE_PATH, "wb") as f2:
        f2.write(f1.read())
    
with sqlite3.connect(DB_FILE_PATH) as conn:
    query = f"select * from {TABLE_NAME}"
    sustainable_product_data = pd.read_sql_query(query, conn)
    print(f"Successfully loaded data from {TABLE_NAME}!")
```
